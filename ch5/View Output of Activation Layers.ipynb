```python
from keras.models import load_model, Model
import matplotlib.pyplot as plt
from imageio import imread
from skimage.transform import resize
import numpy as np
```

```python
model = load_model('cats_and_dogs_small_2.h5')
model.summary()
```

```python
img_tensor = np.array(imread('./train/cats/cat.1.jpg')).astype(float)
img_tensor = resize(img_tensor, (150, 150, 3)) / 255.
img_tensor = np.expand_dims(img_tensor, axis=0) # add another axis
plt.imshow(img_tensor[0], cmap='gray')

print(img_tensor.shape)
```

```python
layer_outputs = [l.output for l in model.layers]
activation_model = Model(inputs=model.input,
                         outputs=layer_outputs)
```

```python
activations = activation_model.predict(img_tensor)
activations[0].shape # output of first layer activation
```

```python
# can view output of different convolution layers (0-7)
LAYER=2

# can visualize every channel by changing color channel 
CHANNEL=2
plt.imshow(activations[LAYER][0,:,:,CHANNEL], cmap='gray')
```
